```python
import torch
from diffusers import StableDiffusionPipeline
from IPython.display import Image as IPyImage, display
from google.colab import files
import os

# Configurar o dispositivo para GPU, MPS ou CPU
if torch.cuda.is_available():
    device = "cuda"
elif torch.backends.mps.is_available():
    device = "mps"
else:
    device = "cpu"

# Carregar o modelo pré-treinado do Stable Diffusion
pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4")
pipe = pipe.to(device)

# Gerar uma imagem a partir de um prompt de texto
prompt = "image with three pigs"
image = pipe(prompt).images[0]

# Salvar a imagem gerada
image.save("generated_image.png")
print("Imagem gerada e salva como 'generated_image.png'.")

# Listar o arquivo salvo para verificar se foi salvo corretamente
!ls -l generated_image.png

# Exibir a imagem gerada no notebook
display(IPyImage(filename="generated_image.png"))

# Se necessário, baixar a imagem
# files.download("generated_image.png")

# Verificar o diretório de trabalho atual
print(f"Diretório de trabalho atual: {os.getcwd()}")
```
