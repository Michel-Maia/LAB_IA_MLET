Tokenização - Dividir o texto em pequenas frases ou palavras para que sejam processadas pela máquina

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
frases = ['Hoje o dia está lindo, mas ontem estava chovendo muito',
          'Ontem fui para a aula de karate, estava chovendo  quando sai de casa',
          'Vamos ao parque hoje ?']
```

```python
CountVec = CountVectorizer()
```

fit_transform
para treinar o modelo

```python
bag_of_words = CountVec.fit_transform(frases)
```

```python
bag_of_words
```

remover os 0 (zero) da matrix sparse

toarray() transforma a matrix sparse em array

get_feature_names_out() obtem os nomes de saída

```python
bow = pd.DataFrame(bag_of_words.toarray(), columns = CountVec.get_feature_names_out())
bow
```
