```python
# Criando um dataset para teste
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Definir a faixa de temperatura e vendas
temperatura_min = 15
temperatura_max = 35
vendas_min = 50
vendas_max = 200

# Gerar dados aleatrios de temperatura
num_amostras = 500  # Nmero de amostras a serem geradas
temperaturas = np.random.rand(num_amostras) * (temperatura_max - temperatura_min) + temperatura_min

# Calcular as vendas de sorvete (funo linear)
def vendas_sorvete(temperatura):
  m = 2    # Inclinao da reta de regresso
  b = 100  # Intercepto da reta de regresso
  return m * temperatura + b

vendas_sorvete = vendas_sorvete(temperaturas)

# Adicionar rudo aleatrio s vendas
vendas_sorvete += np.random.randn(num_amostras) * 10

# Criar o DataFrame
dados = pd.DataFrame({
  "Temperatura": temperaturas,
  "Vendas_Sorvete": vendas_sorvete
})
# Visualizar a distribuio dos dados
plt.scatter(dados["Temperatura"], dados["Vendas_Sorvete"])
plt.xlabel("Temperatura")
plt.ylabel("Vendas de Sorvete")
plt.title("Distribuio dos Dados")
plt.show()
```
