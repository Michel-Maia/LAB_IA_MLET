# Tipos de tokenização


```python
texto = "Olá, mundo! Bem-vindos ao processamento de linguagem natural."
```

```python
# Tokenização por caractere (Character Tokenization)
# divide o texto em caracteres individuais, simples porém não considera a estrutura das palavras

tokens = list(texto)

print(tokens)
```

```python
# Tokenização por palavra (Word Tokenization)
# divide o texto em palavras individuais (biblioteca comum de PLN - nltk natural language toolkit)
import nltk
from nltk import word_tokenize
nltk.download('punkt')

tokens = word_tokenize(texto)
print(tokens)
```

```python
# Tokenização por subpalavra (Subword Tokenization)
# divide o tezto em subpalavras ou fragmentos
# bem-vindo e bem-vindos seria agrupado e mantem a informação de plural com o token ##s
# tokenização por subpalavra também separa conjugações diferentes de verbo, prefixo e sufixo
# mantendo o token raiz e os demais

from transformers import BertTokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
tokens = tokenizer.tokenize(texto)
print(tokens)
```

```python
#transformação de texto para minúsculo
texto = texto.lower()
print(texto)
```
